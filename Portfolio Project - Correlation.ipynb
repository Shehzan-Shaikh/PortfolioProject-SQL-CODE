```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read the data

df = pd.read_csv(r'C:\Users\Shehzan Shaikh\Desktop\Alex The Analyst\movies.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
# Finding the missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col,pct_missing))

# No data is missing
```

```python
# Data Types for our Columns

df.dtypes
```

```python
# Changing the data type of the columns

df['budget'] = df['budget'].fillna(0)
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].fillna(0)
df['gross'] = df['gross'].astype('int64')

#If your DF is big, you're probably not seeing the missing numbers. But you can use the fillna function to help 
```

```python
df.head()
```

```python
# Sorted the data in descending format

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# This option is used to display the whole dataset

pd.set_option('display.max_rows',None)
```

```python
# Drop any duplicate

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# I bellieve BUDGET will have HIGH CORRELATION
# COMPANY will have HIGH CORRELATION
# ScatterPlot with BUDGET vs GROSS

plt.scatter(x=df['budget'] , y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget Earning')
plt.show()
```

```python
# Seaborn pregression plot budget vs gross

sns.regplot(x='budget',y='gross', data=df , scatter_kws={'color':'orange'} , line_kws={'color' : 'Blue'})
```

```python
# Looking at correlation
df.corr()
```

```python
# There are 3 types of correlation
# 1) Pearson is default ,  2) Kendall , 3) Spearman
df.corr(method = 'pearson')
```

```python
# There's high correaltion between budget and gross (0.750 and 1)
correlation_matrix = df.corr(method = 'pearson')
plt.title('Correlation Matrix For Numeric Feature')

sns.heatmap(correlation_matrix , annot = True)

```
